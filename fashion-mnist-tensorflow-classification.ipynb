# libraries and the Database

```python
import numpy as np
import datetime
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
```

# Preprocessing



# Loading database

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

# Normalizing images

We divide each pixel of the images from the training and test bases, using the largest value which is 255

With this, each pixel will be in the range between 0 and 1. This way, the neural network will train faster

```python
X_train = X_train / 255.0
```

```python
X_test = X_test / 255.0
```

```python
X_train[0]
```

# Database reshaping

As we are working with a dense neural network, we changed the dimension of the databases to be in vector format

```python
X_train.shape
```

```python
# As the dimension of each image is 28x28, we changed the entire database to the format [-1 (all elements), height * width]
X_train = X_train.reshape(-1, 28*28)
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
# We also changed the test base dimension
X_test = X_test.reshape(-1, 28*28)
```

```python
X_test.shape
```

# Building the Artificial Neural Network

# Defining the model

We define an object of type Sequential (sequence of layers)

```python
model = tf.keras.models.Sequential()
```

```python
model
```

# Adding the first dense (fully-connected) layer

Layer hyper-parameters:
- number of units/neurons: 128
- activation function: ReLU
- input_shape (input layer): (784, )

```python
model.add(tf.keras.layers.Dense(units=128, activation='relu', input_shape=(784, )))
```

# Adding Dropout

Dropout is a regularization technique in which some neurons in the layer have their value changed to zero, that is, during training these neurons will not be updated. As a result, we have less chance of overfitting

```python
model.add(tf.keras.layers.Dropout(0.2))
```

# Adding the output layer

- units: number of classes (10 in the Fashion MNIST database)
- activation function: softmax

```python
model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
```

# Compiling the model

- Optimizer: Adam
- Loss (error function): Sparse softmax (categorical) crossentropy

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
```

```python
model.summary()
```

# Training the model

```python
model.fit(X_train, y_train, epochs=5)
```

# Model evaluation and forecast

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```

```python
print("Test accuracy: {}".format(test_accuracy))
```
